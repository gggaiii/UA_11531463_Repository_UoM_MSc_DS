```python
import pandas as pd

df = pd.read_csv('C:\\StudySources\\Master\\ERP\\Datasets\\2021\\combined_output.csv')
```

```python
df = df[df['2021 super output area - lower layer'].notna() & (df['2021 super output area - lower layer'] != '')]
```

```python
df.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Remove commas and convert columns to numeric, ignore errors for non-numeric columns
df_clean = df.iloc[:, 2:].apply(lambda x: pd.to_numeric(x.astype(str).str.replace(',', ''), errors='coerce'))

plt.figure(figsize=(10, 8))
sns.heatmap(df_clean.corr(), annot=True, fmt='.2f', cmap='coolwarm', square=True)
plt.title('Covariance Matrix')
plt.show()
```

```python
columns_of_interest = [
    'Work mainly at or from home',
    'Underground, metro, light rail, tram',
    'Train',
    'Bus, minibus or coach',
    'Taxi',
    'Motorcycle, scooter or moped',
    'Driving a car or van',
    'Passenger in a car or van',
    'Bicycle',
    'On foot',
    'Other method of travel to work'
]

for col in columns_of_interest:
    print(f"\nTop 10 correlations for '{col}':")
    display(df_clean.corr()[col].abs().sort_values(ascending=False)[1:11])
```
